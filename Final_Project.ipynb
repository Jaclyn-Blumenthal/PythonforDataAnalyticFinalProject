```python
import collections
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

songdf = pd.read_csv('https://raw.githubusercontent.com/Jaclyn-Blumenthal/PythonforDataAnalyticFinalProject/main/dataset.csv')



# let's look at the data
print(songdf)

print(songdf.shape)
#Remove duplicates from Track_Id
songdf.drop_duplicates(subset=['track_id'], inplace=True)
print(songdf.shape)

# Get duration stats
duration = songdf['duration_ms']
print(duration)

#mean of duration
mean = np.mean(duration)
print(f'Mean: {mean}')

#Median of duration
median = np.median(duration)
print(f'Median: {median}')

#Mode of duration
mode = duration.mode()
print(f'Mode: {mode[0]}')

#for duration outliers can be that some songs have people
#talking in the beginning for a story or just background music without the singing

energy = songdf['energy']
print(energy)

mean = np.mean(energy)
print(f'Mean:{mean}')

median = np.median(energy)
print(f'Median:{median}')

mode = energy.mode()
print(f'Mode: {mode[0]}')

#some songs change the amount of energy throughout the song and not always consistent

popularity = songdf['popularity']
print(popularity)

mean = np.mean(popularity)
print(f'Mean:{mean}')

median = np.median(popularity)
print(f'Median:{median}')

mode = popularity.mode()
print(f'Mode: {mode[0]}')
#information can change because some songs take more time to get noticed than others and
#this information was released on a date when all songs were not available for same amout of time

loudness = songdf['loudness']
print(loudness)

mean = np.mean(loudness)
print(f'Mean:{mean}')

median = np.median(loudness)
print(f'Median:{median}')

mode = loudness.mode()
print(f'Mode: {mode[0]}')

#Loudness can change throughout a song from verse to verse

danceability = songdf['danceability']

mean = np.mean(danceability)
print(f'Mean:{mean}')

median = np.median(danceability)
print(f'Median:{median}')

mode = danceability.mode()
print(f'Mode: {mode[0]}')

#Danceability is more biased there are multiple forms of dance and they may only be considering some forms

#Create a Scatter plot for Danceability and Popularity
fig, ax = plt.subplots() #create a figure & axes objects
ax.plot(songdf['danceability'], songdf['popularity'], color= 'purple', marker='^', markersize=5, alpha=.5, linestyle='none') #plot x and y columns as a scatter plot
ax.set_xlabel('Danceability') #adds labels to x axis
ax.set_ylabel('Popularity')
ax.set_title('Danceability vs Popularity')
plt.show() #display the plot

#Create a scatter plot for Loudness vs Popularity
fig, ax = plt.subplots() #create a figure & axes objects
ax.plot(songdf['loudness'], songdf['popularity'], color= 'green', marker='*', markersize=3, alpha=.75, linestyle='none') #plot x and y columns as a scatter plot
ax.set_xlabel('Loudness') #adds labels to x axis
ax.set_ylabel('Popularity')
ax.set_title('Loudness vs Popularity')
plt.show() #display the plot

#Create a histogram of popularity vs number occurences
fig, ax = plt.subplots()
ax.hist(songdf['popularity'])
ax.set_xlabel('Popularity')
ax.set_ylabel('Number of Occurances')
ax.set_title('Popularity Stats')
plt.show()

#Scatter plot of Energy vs Popularity
fig, ax = plt.subplots() #create a figure & axes objects
ax.plot(songdf['energy'], songdf['popularity'], color= 'pink', marker='^', markersize=4, alpha=.75, linestyle='none') #plot x and y columns as a scatter plot
ax.set_xlabel('Energy') #adds labels to x axis
ax.set_ylabel('Popularity')
ax.set_title('Energy vs Popularity')
plt.show() #display the plot

#Scatter plot for Duration vs Popularity
fig, ax = plt.subplots() #create a figure & axes objects
ax.plot(songdf['duration_ms'], songdf['popularity'], color= 'red', marker='*', markersize=3, alpha=.80, linestyle='none') #plot x and y columns as a scatter plot
ax.set_xlabel('Duration') #adds labels to x axis
ax.set_ylabel('Popularity')
ax.set_title('Duration vs Popularity')
plt.show() #display the plot

#Create histogram of Danceability vs # of occurences
fig, ax = plt.subplots()
ax.hist(songdf['danceability'])
ax.set_xlabel('Danceability')
ax.set_ylabel('Number of Occurances')
ax.set_title('Danceability Stats')
plt.show()

#Create a histogram of Enery vs # of occurences
fig, ax = plt.subplots()
ax.hist(songdf['energy'])
ax.set_xlabel('Energy')
ax.set_ylabel('Number of Occurances')
ax.set_title('Energy Stats')
plt.show()

```
